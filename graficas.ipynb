## ==================================================================
## Gráfico de Barras – 1. Producción de Energía Renovable por Fuente
## ==================================================================

```python
import matplotlib.pyplot as plt
import numpy as np
import json

# Cargar datos
with open('./json/02-consumo-de-energia-renovable-moderna.json', 'r') as f:
    data = json.load(f)

years = [int(d['year']) for d in data]
wind = [float(d['WindGeneration']) for d in data]
solar = [float(d['SolarGeneration']) for d in data]
hydro = [float(d['HydroGeneration']) for d in data]
bio = [float(d['GeoBiomass']) for d in data]

x = np.arange(len(years))
width = 0.2

plt.bar(x - width*1.5, wind, width, label='Eólica', color='skyblue')
plt.bar(x - width/2, solar, width, label='Solar', color='orange')
plt.bar(x + width/2, hydro, width, label='Hidráulica', color='lightgreen')
plt.bar(x + width*1.5, bio, width, label='Bioenergía', color='brown')

plt.xlabel('Año')
plt.ylabel('Teravatios hora (TWh)')
plt.title('Producción de Energía Renovable por Fuente')
plt.xticks(x, years, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## ========================================================================
## Gráfico de Torta – 2. Participación de Energías Renovables (último año)
## ========================================================================

```python
import matplotlib.pyplot as plt
import numpy as np
import json

with open('./json/03-Electricidad-eolica-Electricidad-hidroelectrica-Electricidad-solar-Otras-renovables.json', 'r') as f:
    data = json.load(f)

ultimo = data[-1]

valores = [
    float(ultimo.get('solar', 0)),
    float(ultimo.get('wind', 0)),
    float(ultimo.get('hydro', 0)),
    float(ultimo.get('otherRenewables', 0))
]

labels = ['Solar', 'Eólica', 'Hidráulica', 'Otras Renovables']
colors = ['#4CAF50', '#2196F3', '#FFC107', '#00BCD4']

plt.pie(valores, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Participación de Energías Renovables (%)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# ===================================================================
# Gráfico de Área – 4. Consumo de Energía: Renovable vs Convencional
# ===================================================================

```python
import matplotlib.pyplot as plt
import numpy as np
import json

with open('./json/02-consumo-de-energia-renovable-moderna.json', 'r') as f:
    data = json.load(f)

years = [int(d['year']) for d in data]
renovable = [float(d.get('ModernRenewableEnergy', 0)) for d in data]
convencional = [float(d.get('ConventionalEnergy', 0)) for d in data]

plt.fill_between(years, renovable, label='Energía Renovable', color='green', alpha=0.4)
plt.plot(years, renovable, color='green')

plt.fill_between(years, convencional, label='Energía Convencional', color='red', alpha=0.4)
plt.plot(years, convencional, color='red')

plt.xlabel('Año')
plt.ylabel('Consumo de Energía (TWh)')
plt.title('Comparación entre Energía Renovable y Convencional (TWh)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
